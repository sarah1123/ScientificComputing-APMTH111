```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. append to a list
# 2. set a default value for a function input
# 3. turn a list into an array

def f(x):
    return x**3-25

def bisection(a,b,f,tol=1e-4):
    diff = 2*tol
    runs = 1
    xlist = []
    while diff >= tol and runs < 1000:
        xNew  = (a+b)/2
        diff = a - xNew
        x = xNew
        if f(x) == 0:
            a = x
            b = x
        if f(a)*f(x) > 0:
            a = x
        else:
            b = x
        runs = runs + 1
        xlist.append(x)
    return np.array(xlist)

xkvector = bisection(3,2,f,1e-6)
print(xkvector)
```

```python
plt.plot(xkvector,'.')
plt.show()
plt.plot(xkvector,f(xkvector),'.')
```

```python
# 4. access a subset of elements in a list
# find x_{k+1} - x_k
print(xkvector[1:]-xkvector[0:-1])
print(len(xkvector),len(xkvector[1:]))
```
