```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
rng = np.random.default_rng()
```

```python
# find and plot the error for many different h values for a particular finite difference method
# use forward differences for this example

def dfdt_forward(f,x,h):
    return (f(x+h)-f(x))/h

hvals = h=np.logspace(-14,-1,14)
print(hvals)

# use forward differences on sin(x) at pi/3 with the list of hvals.
deriv_estimates = dfdt_forward(np.sin,np.pi/3,hvals)

# find error
deriv = np.cos(np.pi/3)
errors = np.abs(deriv-deriv_estimates)

# plot
plt.plot(hvals,errors,"o--")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$\log_{10}$h")
plt.ylabel("$\log_{10}$error")
plt.title("log-log plot of error")
```
