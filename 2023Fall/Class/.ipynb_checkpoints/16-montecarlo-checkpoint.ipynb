```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(12345)
```

```python
def f(x):
     return np.exp(x) - 2*np.sin(4*x) + 1

num_points = 4
repeat = 1000
MC_list = []
for k1 in range(repeat):
    # use domain of [0,2]
    xvalues = rng.uniform(size=num_points,low=0,high=2)
    domain_size = 2
    MC_list.append(domain_size*np.mean(f(xvalues)))
plt.figure(figsize=[3,2])
plt.hist(MC_list)
other_interp = [8.439, 7.797, 7.642]
[plt.plot([k,k],[0,175]) for k in other_interp]
plt.plot([7.816, 7.816],[0,175],'k')
plt.xlabel('Monte Carlo estimate')
plt.ylabel('count')
print(np.std(MC_list))
plt.savefig('C16-MC4pts.png',dpi=150,bbox_inches = 'tight')
```

```python
# Implement the composite trapezoid rule
# f: the function to integrate 
# xd: the integration points 
def comp_trap(f,xd):
    n=len(xd)
    result=0
    result+=f(xd[0])
    for i in range(n-2):
        result+=2*f(xd[i+1])
    result+=f(xd[-1])
    result=result*(xd[1]-xd[0])/2

    return result

def gauss2(interval,f,k,sp):
    p = interval[1]-interval[0]
    h = p/k
    tot = 0
    for c in range(k):
        x0 = a + h-h/np.sqrt(3)/2
        x1 = a + h+h/np.sqrt(3)/2
        f0 = f(x0)
        f1 = f(x1)
        tot = tot + (f0 + f1)*h/2
        xp = np.arange(c*h,c*h+h,0.001)
        sp.plot(xp,f(xp),color="blue")
        sp.plot(xp,f(x0)+(f1-f0)*(xp-x0)/(x1-x0),color='black')
    return tot

def mid(interval,f,k,sp):
    a = interval[0]
    h = interval[1]-interval[0]
    tot = 0
    x00 = a + h/k/2
    for c in range(k):
        x0 = x00+c*h/k
        f0 = f(x0)
        tot = tot + f0*h/k

        xp = np.arange(x0-x00,x0+x00,0.01)
        sp.plot(xp,f(x0)+0*xp,color='black')
        sp.plot(xp,f(xp),color='blue')
    return tot



# Implement the chi function 
def chi(xd):
    if np.linalg.norm(xd)<1:
        return 1
    else:
        return 0
```

```python
k1 = 5
k2 = 100
est = np.zeros([k1*k2,2])
sigmaf = np.zeros([k1*k2,2])
counter = 0
for k in range(k1):
    for j in range(k2):
        npoints = 10**(k+2)
        xv = rng.uniform(0,1,npoints)
        yv = rng.uniform(0,1,npoints)
        # set to 1 if in domain; 0 if not
        f = (xv**2+yv**2)<1
        sigmaf[counter,:] = [k+2,np.std(f)]
        est[counter,:] = [k+2,4*np.mean(f)]
        counter = counter+1

```

```python
import matplotlib.pyplot as plt
plt.plot(est[:,0],est[:,1],'.')
plt.plot(sigmaf[:,0],3.14-4*sigmaf[:,1]/np.sqrt(10**sigmaf[:,0]),'k')
plt.plot(sigmaf[:,0],3.14+4*sigmaf[:,1]/np.sqrt(10**sigmaf[:,0]),'k')
plt.xlim([1.8,6.2])
plt.ylim([2.5,3.7])
plt.xlabel('log10(N)');
plt.ylabel('estimate of pi');
plt.savefig('C17piest.png',bbox_inches='tight')
```
