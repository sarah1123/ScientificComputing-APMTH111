```python
"""plot sqrt(x) and a linear approximation to it"""
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,4,100)
x2 = np.linspace(1,4,3)

# the two functions (sqrt(x) and the linear approx).
y = np.sqrt(x)
y2 = 1 + (x2-1)/3

fig = plt.figure(figsize=(3, 2)) 
ax = fig.add_subplot(1, 1, 1)

# set up the frame / axis line positions
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.ylim(0,2)
plt.xlabel('x',loc='right')
plt.ylabel('y',loc='top',rotation=0)

# plot the function
plt.plot(x,y, 'k')
plt.plot(x2,y2, 'b')

# show or save the plot
# plt.show()
plt.savefig("C01linearsqrtapprox.png", bbox_inches="tight")
```

```python
"""plot sin(x) and the condition number.  Use both relative and absolute error."""
import matplotlib.ticker as tck
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,3*np.pi,100)
x2 = np.linspace(0.01,3*np.pi,100)
y = np.sin(x)
y2 = np.abs(((x2*np.cos(x2))/np.sin(x2)))

# plot the function
fig = plt.figure(figsize=(6, 2)) 
ax = fig.add_subplot(1, 2, 1)
# 'zero', center' are two options for axis position
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.ylim(-1.5,1.5)
plt.xlabel('x',loc='right')
plt.ylabel('y',loc='top',rotation=0)
plt.plot(x/np.pi,y, 'k')
# adjust the tick labels
ax.xaxis.set_major_locator(tck.MultipleLocator(base=1))
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))

# plot the condition number
ax = fig.add_subplot(1, 2, 2) # 2nd axes
ax.spines['left'].set_position('zero')
# 'zero', center' are two options for axis position
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.ylim(-0.5,20)
plt.xlabel('x',loc='right')
plt.ylabel('y',loc='top',rotation=0)
# need handles to the plots for the legend
line1, = plt.plot(x2/np.pi,y2, 'k', label='rel err')
line2, = plt.plot(x2/np.pi,np.abs(np.cos(x2)), 'b--', label='abs err')
ax.legend(handles=[line1, line2])
ax.xaxis.set_major_locator(tck.MultipleLocator(base=1))
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))

# show or save the plot
# plt.show()
plt.savefig("C01sin.png", bbox_inches="tight")
```
